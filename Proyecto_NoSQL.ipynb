```python
pipeline = [
    {"$match": {"_id": 1}},

    {"$unwind": "$productes"},

    {"$lookup": {
        "from": "productes",
        "pipeline": [
            {"$unwind": "$productes"},
            {
                "$match": {
                    "$expr": {
                        "$eq": [
                            {"$toInt": "$productes.id"},
                            "$$id_prod"
                        ]
                    }
                }
            }
        ],
        "let": {"id_prod": "$productes.id_producte"},
        "as": "detall_producte"
    }},

    {"$unwind": "$detall_producte"},
    {"$replaceRoot": {"newRoot": "$detall_producte.productes"}},

    {"$unwind": "$ingredients"},

    {"$addFields": {
        "allergen": {
            "$switch": {
                "branches": [
                    {
                        "case": {
                            "$regexMatch": {
                                "input": "$ingredients.nom",
                                "regex": "formatge|llet|nata|gelat",
                                "options": "i"
                            }
                        },
                        "then": "Lactis"
                    },
                    {
                        "case": {
                            "$regexMatch": {
                                "input": "$ingredients.nom",
                                "regex": "pa",
                                "options": "i"
                            }
                        },
                        "then": "Gluten"
                    }
                ],
                "default": "Sense al·lèrgens comuns"
            }
        }
    }},

    {"$project": {
        "_id": 0,
        "producte": "$nom_producte",
        "ingredient": "$ingredients.nom",
        "allergen": 1
    }}
]
```

```python
resultat = list(db.comandes.aggregate(pipeline))

for r in resultat:
    print(r)
```

Consultar preu mig d'un producte a partir del seu id

```python
id_prod = 4
filtre = [
    {
        "$unwind": "$productes"
    },
    {
        "$match": {
            "productes.id_producte": id_prod,
            "productes.valoracio": { "$ne": None }
        }
    },
    {
        "$group": {
            "_id": "$productes.id_producte",
            "preu_mig": { "$avg": "$productes.valoracio" }
        }
    }
]

resultat = list(db.comandes.aggregate(filtre))

if resultat:
    preu_mig = resultat[0]["preu_mig"]
    print(f"El preu mig del producte {id_prod} és: {preu_mig:.2f}")
else:
    print(f"No s´han trobat valoracions pel producte {id_prod}.")
```

```python
nom_producte = "Hamburguesa Clàssica"
producte = db.productes.find_one({"nom_producte": nom_producte})

total_calories = 0
total_proteines = 0
total_carbohidrats = 0
total_greixos = 0

if producte and "ingredients" in producte:
    print(f"Calculant informació nutricional per: {nom_producte}")
    for ingredient_info in producte["ingredients"]:
        ingredient_id = ingredient_info["id_ingredient"]

        quantity = float(ingredient_info.get("quantitat", 1))

        subministre = db.subministres.find_one({"id_ingredient": ingredient_id})

        if subministre and "nutrientes" in subministre:
            print(f"  Ingredient: {subministre['nom']}, Quantity: {quantity}")
            for nutrient in subministre["nutrientes"]:
                if nutrient["tipo"] == "proteines":
                    total_proteins += nutrient["grams"] * quantity
                elif nutrient["tipo"] == "hidrats":
                    total_carbohydrates += nutrient["grams"] * quantity
                elif nutrient["tipo"] == "greixos":
                    total_fats += nutrient["grams"] * quantity

            total_calories += subministre['calories'] * quantity
        else:
            print(f"  No s'ha trobat informació nutricional per l'ingredient: {ingredient_id}")
else:
    print(f"El producte '{product_name}' no s'ha trobat o no té ingredients.")

print(f"\nInformació total per a {product_name}:")
print(f"  Total Calories: {total_calories:.2f} kcal")
print(f"  Total Proteines: {total_proteins:.2f} grams")
print(f"  Total Carbohidrats: {total_carbohydrates:.2f} grams")
print(f"  Total Greixos: {total_fats:.2f} grams")
```

```python

proveidors = list(db.proveidors.find({},
             {"_id" : 0, "id_proveidor": 1, "nom" : 1}))

resultats = []
for p in proveidors
  compres = list(db.compra.find(
      {"id_proveidor": p["id_proveidor"]},
      {"_id": 0, "id_compra": 1}))
  total_vendes = 0
  for c in compres
    detalls = list(db.compra_proveidor.find(
        {"id_compra": c["id_compra"]},
        {"_id": 0, "cost_total": 1}))
    for d in detalls
      total_vendes += d["cost_total"]

  resultats.append({"Proveidor": p["nom"],
                    "Total_vendes": total_vendes})

resultats.sort(key=lambda x: x["Volum_vendes"], reverse=True)
```

```python
df = pd.DataFrame(resultats)
display(df)
```

```python
proveidors = list(db.proveidors.find({},
              {"_id": 0, "id_proveidor": 1, "nom": 1}))
for prov in proveidors:
    nom_prov = prov["nom"]
    id_prov = prov["id_proveidor"]


    compres = list(db.compra.find({"id_proveidor": id_prov},
                                  {"_id": 0, "id_compra": 1}))
    for c in compres:

        detalls = list(
            db.compras_proveidor.find(
                {"id_compra": c["id_compra"]},
                {"_id": 0, "id_producte": 1, "quantitat": 1}
            )
        )
        for d in detalls:
            id_prod = d["id_producte"]

            subm = db.subministres.find_one(
                {"id_producte": id_prod}, {"_id": 0, "nom": 1}
            )

            subprov = db.proveidor_subministres.find_one(
                {"id_proveidor": id_prov, "id_producte": id_prod},
                {"_id": 0, "quantitat_unitat": 1, "unitats": 1}
            )

            rows.append({
                "Proveidor": nom_prov,
                "Producte": subm["nom"] if subm else None,
                "quantitat": d["quantitat"],
                "quantitat_unitat": subprov["quantitat_unitat"] if subprov else None,
                "unitats": subprov["unitats"] if subprov else None
            })
```
